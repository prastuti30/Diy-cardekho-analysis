```python
#importing essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#inporting our csv file 
data = pd.read_csv("car_dekho.csv")
data.head()
data.tail()
```

The file is being fetched successfully.!

```python
#checking the statistical features of the data
data.describe()
```

describe() function provides the details of the dataset. Based on this, we can answer the questions based on records like total sales, minimum or maximum sales, missing records, average sales, etc.


1. The data here contains cars with manufacturing years 2003 to 2018.
2. The lowest price for the car that was sold is 10,000 INR.
3. The highest price at which the car was sold is 35 lakh INR.
4.  There are altogether 301 records in the data shown by count().

```python
data.info() #helps to look at different variables present in the dataset
```

```python
print(data.Car_Name.value_counts())
      #checking whether non numeric variables are categorical in nature or not
# used to find the car types
```

```python
fig_dims = (12,20)
fig, ax = plt.subplots(figsize = fig_dims)
sns.countplot(y= data.Car_Name, ax = ax, data = data)
        
```

The matplotlib function subplots() plot the bar graph of Car Name v/s count 

```python
print(data.Fuel_Type.value_counts())
```

gives the fuel type of the different car available on the platform

```python
plt.pie(data.Fuel_Type.value_counts(), labels=["PETROL","DIESEL","CNG"],shadow = False, autopct='%1.1f%%')
plt.title("fuel distribution")
```

```python
data.Fuel_Type.value_counts().plot.bar()
```

```python
#selling price vs present price
#ploting a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Present_Price'], data['Selling_Price'], alpha=0.5)
plt.title('Ploting a Scatter Plot of Selling Price vs Present Price')

plt.xlabel('Present Price (in Lakhs)')
plt.ylabel('Selling Price (in Lakhs)')
plt.grid(True)
plt.show()
```

```python
# Histogram for Kms Driven distribution
plt.figure(figsize=(10, 6))
plt.hist(data['Kms_Driven'], bins=30, edgecolor='k', alpha=0.7)
plt.title('Ploting a histogram of Kms Driven Distribution')
plt.xlabel('Kms Driven')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
year_counts = data['Year'].value_counts().sort_index()
plt.bar(year_counts.index, year_counts.values, color='skyblue')
plt.title('Count of Cars by Year')
plt.xlabel('Year')
plt.ylabel('Number of Cars')
plt.grid(axis='y')
plt.show()
```

```python
# Bar plot for Transmission Type distribution
plt.figure(figsize=(8, 6))
transmission_counts = data['Transmission'].value_counts()
plt.bar(transmission_counts.index, transmission_counts.values, color='lightgreen')
plt.title('Transmission Type Distribution')
plt.xlabel('Transmission Type')
plt.ylabel('Number of Cars')
plt.grid(axis='y')
plt.show()
```
